```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#initialising the file
file = pd.read_csv("D:\\PDFs\\CSV Files\\spotify_data clean.csv")
f = pd.DataFrame(file)
file2 = pd.read_csv("D:\\PDFs\\CSV Files\\track_data_final.csv")
f2 = pd.DataFrame(file2)
```

```python
#seeing how many explicit songs are there
sns.countplot(data=f, x='explicit')
plt.show()
```

```python
#seeing how many artists there are in the list and how many have come repeatedly
singers = f.groupby('artist_name').agg(counts=('artist_name','count')).sort_values(by='counts', ascending=False)
print(singers)
print(f"Top singer is {singers.index[0]}")
```

```python
#seeing the popularity of TS tracks visually
taylor_songs = f[f['artist_name'] == singers.index[0]].sort_values(by='track_popularity', ascending=False)
sns.scatterplot(data=taylor_songs, y='track_popularity', x='track_duration_min')
plt.show()
```
